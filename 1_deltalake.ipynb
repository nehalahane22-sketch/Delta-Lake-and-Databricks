# Create delta table

### **Using sql API**

```python
%sql
CREATE TABLE databricksnl.default.first_table
(
  id INT,
  salary INT
)
USING DELTA
```

### DELTA USING api

```python
from delta.tables import DeltaTable
```

```python
DeltaTable.createIfNotExists(spark) \
  .tableName("databricksnl.default.firstdeltaapi") \
  .addColumn("id", "INT") \
  .addColumn("salary", "INT") \
  .execute()
```

```python
from pyspark.sql.functions import *
from delta.tables import DeltaTable, IdentityGenerator
from pyspark.sql.types import LongType ,IntegerType, StringType
```

```python
%sql
drop table databricksnl.default.firstdeltaapi
```

```python


DeltaTable.create(spark) \
  .tableName("databricksnl.default.firstdeltaapi") \
  .addColumn("salarafter_Tax", dataType=LongType(), generatedAlwaysAs="CAST((salary * 0.7) AS BIGINT)") \
  .addColumn("salary", dataType=IntegerType()) \
  .addColumn("name", dataType=StringType()) \
  .execute()
```
