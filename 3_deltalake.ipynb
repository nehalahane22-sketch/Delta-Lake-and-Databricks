# schema enforcment

```python
data = [(1,100,'aa',1),(2,200,'bb',1),(3,300,'cc ',1),(4,400,'dd ',1)]
df = spark.createDataFrame(data,['id','salary','name','sus_col'])

display(df)
```

```python
df.write.format("delta")\
    .mode("append")\
    .save('/Volumes/databricksnl/default/deltavol/demosink/')
```

## schema evolution

```python
df.write.format("delta")\
    .mode("append")\
    .option("mergeSchema", "true")\
    .save('/Volumes/databricksnl/default/deltavol/demosink/')
```

schema overwite

```python
data = [(1,100,'aa',1),(2,200,'bb',1),(3,300,'cc ',1),(4,400,'dd ',1)]
df = spark.createDataFrame(data,['cust_id','income','name','tip'])

display(df)
```

```python
df.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema","true")\
    .save('/Volumes/databricksnl/default/deltavol/demosink/')
```
