# delta log level schema changes

```python
data = [(1,100,'xyz',1),(9,200,'bb',1),(10,300,'cc ',1)]
df = spark.createDataFrame(data,['cust_id','income','name','tip'])

display(df)
```

```python
df.write.format("delta")\
    .mode("append")\
    .save('/Volumes/databricksnl/default/deltavol/schemalevel/')
```

```python
df = spark.read.format("parquet")\
    .load("/Volumes/databricksnl/default/deltavol/schemalevel/part-00000-841eabe5-dbc0-4d12-a7d1-4660f9dceaad.c000.snappy.parquet")
display(df)
```

```python
%sql
ALTER TABLE delta.`/Volumes/databricksnl/default/deltavol/schemalevel`
SET TBLPROPERTIES (
  'delta.columnMapping.mode' = 'name',
  'delta.minReaderVersion' = '2',
  'delta.minWriterVersion' = '5'
);


ALTER TABLE delta.`/Volumes/databricksnl/default/deltavol/schemalevel`
RENAME COLUMN name TO customer_name
```
